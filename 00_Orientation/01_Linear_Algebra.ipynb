```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Vector addition
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
v_sum = v1 + v2
print("Vector Sum:", v_sum)

# Dot product
dot = np.dot(v1, v2)
print("Dot Product:", dot)
```

```python
# Matrix creation
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Matrix multiplication
C = np.dot(A, B)
print("Matrix Multiplication:\n", C)

# Matrix inverse
inv_A = np.linalg.inv(A)
print("Inverse of A: \n", inv_A)
```

## Conclusion
- Linear Algebra is essential for ML
- Vectors, matrices, dot products, and inverses are widely used
- Eigenvalues/eigenvectors are important for PCA and other algorithms
