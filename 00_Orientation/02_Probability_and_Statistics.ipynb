# Probability and Statistics for Machine Learning

## Overview
Statistics and probability are essential for understanding data patterns,
making predictions, and evaluating ML models.

## Applications
- Understanding data distribution
- Bayesian algorithms
- Model evaluation
- Feature selection


```python
import numpy as np
import pandas as pd
from scipy import stats

import matplotlib.pyplot as plt
```

## Descriptive Statistics


```python
data = [10, 20, 20, 30, 40, 50, 50, 50, 60]

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True).mode[0]
std_dev = np.std(data)
variance = np.var(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Standard Deviation:", std_dev)
print("Variance:", variance)
```

## Probability Basics
- Probability = Favorable outcomes / Total outcomes


```python
# Example: Coin toss probability
total_outcomes = 2
favourable_outcome = 1
prob = favourable_outcome / total_outcomes
print("Probability of Heads:", prob)
```

## Conditional Probability
P(A|B) = P(A âˆ© B) / P(B)


```python
# Example: simple conditional probability
P_A_and_B = 0.1
P_B = 0.2
P_A_given_B = P_A_and_B / P_B
print("P(A|B):", P_A_given_B)
```

```python
## Visualization
plt.hist(data, bins=6, color='skyblue', edgecolor='black')
plt.title("Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## Conclusion
- Mean, median, mode describe data central tendency
- Variance and standard deviation show spread
- Probability and conditional probability are fundamental for ML algorithms
