# Discrete Math for Machine Learning

## Overview
Discrete Math is important for understanding:
- Logic in algorithms
- Sets for data grouping
- Combinatorics for probability
- Graphs in ML (optional)

## Applications
- Feature selection
- Understanding data structures
- Algorithm design
- Probability calculations


```python
# Sets
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
# Union
print("A ∪ B:", A | B)

# Intersection
print("A ∩ B:", A & B)

# Difference
print("A - B:", A - B)

# Symmetric Difference
print("A △ B:", A ^ B)
```

## Logic Basics
- AND, OR, NOT operations
- Conditional statements


```python
# Logical operations
a = True
b = False

print("a AND b:", a and b)
print("a OR b:", a or b)
print("NOT a:", not a)
```

## Combinatorics
- Factorial, permutations, combinations
- Used in probability and ML


```python
import math
# factorial
print("5! =", math.factorial(5))

#permutations: nPr = n! / (n - r)!
n, r = 5,2
print("5p2 =", math.perm(n, r))

#combinations: nCr = n! / (r! * (n - r)!)
print("5C2 =", math.comb(n, r))
```

## Probability using Combinatorics
- Example: Probability of drawing 2 aces from a deck of 52 cards


```python
# Total ways to choose 2 cards from 52
total_ways = math.comb(52, 2)

#ways to choose 2 aces from 4
favourable = math.comb(4, 2)

probability = favourable / total_ways
print("Probability of drawing 2 aces:", probability)
```
