# Calculus for Machine Learning

## Overview
Calculus helps understand how models learn by optimizing a loss function.
This notebook covers derivatives, partial derivatives, and gradient descent.

## Applications
- Linear Regression optimization
- Neural Network backpropagation
- Logistic Regression


```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Derivative of a Function

- Measures rate of change of a function
- f'(x) = dy/dx


```python
# Example: f(x) = x^2
def f(X):
  return X**2
def df(X):
  return 2*X

X = np.array([0, 1, 2, 3, 4, 5])
y = f(X)
dy = df(X)

print("f(X):", y)
print("f'(X):", dy)
```

## 2. Partial Derivatives

- Measures change of a multivariable function with respect to one variable
- Used in gradient descent


```python
# Example: f(x, y) = x^2 + y^2
def f(x, y):
  return x**2 + y**2
def df_dx(x, y):
  return 2*x
def df_dy(x, y):
  return 2*y
x, y = 3, 4
print("∂f/∂x:", df_dx(x, y))
print("∂f/∂y:", df_dy(x, y))
```

## 3. Gradient Descent

- Algorithm to minimize a function (e.g., loss function)
- Update rule: θ = θ - learning_rate * gradient


```python
# Simple example: f(x) = (x-3)^2
x = 0 #initial value
learning_rate = 0.1
epochs = 20

for i in range(epochs):
  grad = 2*(x - 3)
  X = x - learning_rate * grad
  print(f"Epoch {i+1}: x = {x}, f(x) = {(x-3)**2}")
```

## 4. Visualizing Gradient Descent



```python
x_vals = np.linspace(-1, 6, 100)
y_vals = (x_vals - 3)**2
plt.plot(x_vals, y_vals, label='f(x) = (x-3)^2')
plt.scatter(x, (x-3)**2, color='red', label='Minimum Found')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

## Conclusion

- Derivatives show rate of change
- Partial derivatives are used for multivariable functions
- Gradient descent is the core optimization algorithm in ML
- Understanding calculus helps explain how models learn
