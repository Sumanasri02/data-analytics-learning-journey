#Python Basics for Data Analytics

This notebook covers Python fundamentals required for data analysis.

Code topics

Variables & data types

Lists, tuples, dictionaries

Loops

Functions

```python
# Integer
age = 22

# Float
salary = 45000.75

# String
name = "Sumanasri"

# Boolean
is_student = True

print(type(age))
print(type(salary))
print(type(name))
print(type(is_student))
```

```python
# List (mutable)
scores = [80, 85, 90, 95]
print(scores)
print("First score:", scores[0])

# Tuple (immutable)
dimensions = (1920, 1080)
print(dimensions)

# Dictionary (key-value pairs)
student = {
    "name": "Sumanasri",
    "age": 22,
    "course": "Data Analytics"
}

print(student)
print("Student Name:", student["name"])
```

```python
scores = [80, 85, 90, 95]

for score in scores:
    print(score)
```

```python
count = 1
while count <= 5:
  print("Count:", count)
  count += 1
```

```python
def calculate_average(numbers):
  return sum(numbers) / len(numbers)

scores = [80, 85, 90, 95]
average_score = calculate_average(scores)

print("Averages_score:", average_score)
```
