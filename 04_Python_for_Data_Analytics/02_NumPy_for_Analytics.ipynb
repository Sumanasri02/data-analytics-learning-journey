# NumPy for Data Analytics

NumPy is the core library for numerical computing in Python.
It provides fast operations on arrays and is the foundation for Pandas and ML libraries.


```python
import numpy as np
data = np.array([10, 20, 30, 40, 50])
print(data)
print(type(data))
```

```python
print("Shape:", data.shape)
print("Size:", data.size)
print("Data types:", data.dtype)
```

```python
data = np.array([10, 20, 30, 40, 50])

print("First element:", data[0])
print("Last element:", data[-1])
print("Slice:", data[1:4])
```

```python
## Working with 2D Arrays
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(matrix)
print("Shape:", matrix.shape)
```
