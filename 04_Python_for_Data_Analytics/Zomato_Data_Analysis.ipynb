##Zomato Data Analysis Using Python

## Objective

The objective of this project is to analyze restaurant data from Zomato to understand:

Restaurant types and customer preferences

Online vs offline ordering trends

Rating distribution

Cost preferences for couples

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("/content/Zomato-data-.csv")
print(dataframe.head())
```

Key Columns:

name – Restaurant name

rate – Customer rating

votes – Number of customer votes

online_order – Online order availability

listed_in(type) – Restaurant category

approx_cost(for two people) – Approximate dining cost

Step 3: Data Cleaning

Converted ratings from text format (e.g., 4.2/5) to numeric values.

Checked for missing and null values.

Ensured consistency across columns.

```python
def handleRate(value):
  value = str(value).split('/')
  value = value[0]
  return float(value)
dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
print(dataframe.isnull().sum())
```

```python
sns.countplot(x = dataframe['listed_in(type)'])
plt.xlabel("Type of restaurat")
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c='green', marker='o')
plt.xlabel('Type of restaurant')
plt.ylabel('Votes')
```

```python
max_votes = dataframe['votes'].max()
restaurant_with_max_votes = dataframe.loc[dataframe['votes'] == max_votes, 'name']

print('Restaurant(s) with the maximum votes:')
print(restaurant_with_max_votes)
```

```python
sns.countplot(x=dataframe['online_order'])
```

```python
plt.hist(dataframe['rate'], bins = 5)
plt.title('Ratings Distribution')
plt.show()
```

```python
plt.figure(figsize= (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed In (Type)')
plt.show()
```
